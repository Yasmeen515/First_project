```python
from sklearn.model_selection import train_test_split 
 
train_set, test_set = train_test_split(diamonds, test_size=0.2, random_state=42) 
train_set_price = train_set["price"]
train_set = train_set.drop(['price'],axis=1)
test_set_price = test_set["price"]
test_set = test_set.drop(['price'],axis=1) 
```

```python
#select a model 

from sklearn.linear_model import LinearRegression 

num_pipeline = Pipeline([
('imputer', SimpleImputer(strategy="median")),
('std_scaler', StandardScaler()),
]) 

 

lin_reg = LinearRegression()
lin_reg.fit(train_set, train_set_price) 


test_accuracy = round(lin_reg.score(test_set, test_set_price) * 100, 3)
test_accuracy 




```

```python
from sklearn.metrics import mean_squared_error
lin_diamonds_predictions = lin_reg.predict(test_set)
lin_mse = mean_squared_error(test_set_price, lin_diamonds_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(train_set, train_set_price) 

tree_predictions = tree_reg.predict(test_set)
test_accuracy = round(tree_reg.score(test_set, test_set_price) *100,3)
 
tree_mse = mean_squared_error(test_set_price, tree_predictions)
tree_rmse = np.sqrt(tree_mse) # standard deviation 
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(tree_rmse)
print(tree_rmse) 
print(test_accuracy)
```

```python
print("pedictions:",tree_predictions[:10]) 
print("label:" ,list(test_set_price[:10]))
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(train_set, train_set_price)

forest_diamonds_predictions = forest_reg.predict(test_set) 
forest_mse = mean_squared_error(test_set_price, forest_diamonds_predictions)
forest_rmse = np.sqrt(forest_mse) 

test_accuracy = round(forest_reg.score(test_set, test_set_price) *100 ,3)
display_scores(tree_rmse)
print(test_accuracy)
print(forest_rmse ) 
```

```python
#fine tune

from sklearn.model_selection import RandomizedSearchCV
randomforest = RandomForestRegressor()
hyperparameters = {'bootstrap':[True,False],
                  'n_estimators':[200, 400, 600],
                  'max_features':[3,4],
                  'max_depth':[60,70],
                  'min_samples_leaf':[1],
                  'min_samples_split':[5, 8]}

r_randomized = RandomizedSearchCV(randomforest, param_distributions= hyperparameters, n_iter=48, cv= 3)
r_randomized.fit(train_set, train_set_price)
randomized_predictions = r_randomized.predict(test_set)
test_accuracy = round(r_randomized.score(test_set, test_set_price) * 100, 3)
print(test_accuracy)

```
